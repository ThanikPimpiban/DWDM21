````python
merged_table = flights.merge(airports,how='left',left_on='รหัสสนามบินต้นทาง',right_on='รหัสสนามบิน')  #ทำการเชื่อมตาราง 2 ตาราง ได้แก่ 1.flights 2.airports ด้วย 1.รหัสสนามบินต้นทาง 2.รหัสสนามบิน ตามลำดับ
````

````python
merged_table.dropna() ##เช็คดู missing
````

1.หารัฐที่มีจำนวนเครื่องบินขาออกมากที่สุดและน้อยที่สุด

````python
merged_table.groupby('รัฐ').count()  ### ตรวจดูจำนวนเที่ยวบินของแต่ละรัฐ
````

````python
max = merged_table.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False)  ## Group column รัฐ ที่มีสนามบินปลายทาง มากสุด เพราะโจทย์ถามขาออก
max.iloc[:1,4:5]
````

````python
min = merged_table.groupby('รัฐ').count().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False) ## Group column รัฐ ที่มีสนามบินปลายทาง น้อยสุด เพราะโจทย์ถามขาออก
min.iloc[-1:,4:5]
````

**4.สายการบินอะไรมีความล่าช้ามากที่สุดและน้อยที่สุด**

````python
merged_table['ความล่าช้าขาออก'].max() # ตรวจดู column ความล่าช้าขาออก ที่มีค่ามากสุด
````

````python
merged_table['ความล่าช้าขาเข้า'].max() # ตรวจดู column ความล่าช้าขาออก ที่มีค่ามากสุด
````

````python
#### ตอบความล่าช้าขาออกมากสุด
max2_1 = merged_table.groupby('สายการบิน').max().sort_values(by=['ความล่าช้าขาออก'],ascending=False) ## Group column สายการบิน ที่มีความล่าช้าขาออก มากสุด
max2_1.iloc[:1,4:5]
````

````python
#### ตอบความล่าช้าขาเข้ามากสุด
max2_2 = merged_table.groupby('สายการบิน').max().sort_values(by=['ความล่าช้าขาเข้า'],ascending=False) ## Group column สายการบิน ที่มีความล่าช้าขาออก มากสุด
max2_2.iloc[:1,5:6]
````

````python
merged_table['ความล่าช้าขาออก'].min() # ตรวจดู column ความล่าช้าขาออก ที่มีค่าน้อยสุด
````

````python
merged_table['ความล่าช้าขาเข้า'].min() # ตรวจดู column ความล่าช้าขาเข้า ที่มีค่าน้อยสุด
````

````python
#### ตอบความล่าช้าขาออกน้อยสุด
min2_1 = merged_table.groupby('สายการบิน').min().sort_values(by=['ความล่าช้าขาออก'],ascending=False) # Group column สายการบิน ที่มีความล่าช้าขาออก น้อยสุด
min2_1.iloc[-1:,4:5]
````

````python
#### ตอบความล่าช้าขาเข้าน้อยสุด
min2_1 = merged_table.groupby('สายการบิน').min().sort_values(by=['ความล่าช้าขาเข้า'],ascending=False) # Group column สายการบิน ที่มีความล่าช้าขาเข้า น้อยสุด
min2_1.iloc[-1:,5:6]
````
