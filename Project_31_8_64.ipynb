```python
merged_data_rst_data_st = data_rst.merge(data_st,how='left',left_on='OBJECTID',right_on='id')
merged_data_rst_data_st
```

```python
merged_3_tables =merged_data_rst_data_st.merge(data_t,how='right',left_on='station_code',right_on='StStart')
merged_3_tables
```
