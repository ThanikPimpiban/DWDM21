```python
data_st = pd.read_csv(os.path.join(path,'data_station.csv')) #กำหนดตัวแปร data_st เป็นข้อมูลของสถานีรถไฟแต่ละสถานี
data_st #เรียกดูตารางของสถานีรถไฟแต่ละสถานี
```

```python
data_t = pd.read_csv(os.path.join(path,'data_train.csv')) #กำหนดตัวแปร data_t เป็นข้อมูลของรถไฟ
data_t #เรียกดูตาราง
```

```python
data_rst = pd.read_csv(os.path.join(path,'railstation_new.csv')) #กำหนดตัวแปร data_rst ให้เป็นรางรถไฟ
data_rst #เรียกดูตารางของรางรถไฟ
```

```python
merged_data_rst_data_st = data_rst.merge(data_st,how='left',left_on='NAMET',right_on='name') #ทำการเชื่อมตาราง 2 ตาราง คือ 1.ข้อมูลของรางรถไฟ 2.ข้อมูลของสถานีรถไฟแต่ละสถานี
merged_data_rst_data_st #เรียกตารางที่เชื่อมกัน 2 ตารางข้างต้น
```

```python
merged_3_tables =merged_data_rst_data_st.merge(data_t,how='right',left_on='station_code',right_on='StStart') #ทำเชื่อมตารางระหว่าง 1.ตารางของรางรถไฟและตารางสถานีรถไฟแต่ละสถานี 2.ตารางรถไฟ
merged_3_tables #เรียกดูตารางที่เชื่อมกันทั้ง 3 ตาราง
```

```python
merged_3_tables.isnull().any() #เช็คดู missing ของตารางที่เชื่อมกันทั้ง 3 ตาราง
```

**กำจัดค่า Missing**

```python
# หา mode
merged_3_tables_Nomissing =merged_3_tables[['NAMET']].mode()
# เติม mode ลงไปแทนค่า missing
merged_3_tables[['NAMET']] = merged_3_tables[['NAMET']].fillna('NAMET')
```

```python
merged_3_tables.isnull().any() #เช็ดดู missing ของตารางที่เชื่อมกันทั้ง 3 ตาราง อีกรอบ
```
